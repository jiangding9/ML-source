```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, 5)  #在0-20的长度上获取5个等距的点。
print(x)
y = 2 * x + 1
```

```python
plt.scatter(x, y)  #画图
```

```python
plt.scatter(x, y, s=200, marker = ',')
```

```python
plt.scatter(x, y, s=200, marker = 'v', c = 'y')
plt.show()
```

```python
plt.xlim((0,30))
plt.xlim((0,50))

plt.scatter(x, y)
```

```python
x = np.linspace(-10,10,40)   #这里如果用arange来构建x的话，得到的是连起来的折线图，不是平滑的曲线图
y = x  ** 2 + 1
plt.plot(x, y)
```

```python
x = np.linspace(-10,10,40)   #这里的x不是把x周的标签按分成这么多点。标签需要其他的命令修改
y = x  ** 2 + 1
y0 = -x * x
plt.scatter(x, y)
plt.plot(x, y0, c = 'r')
```

```python
x = np.linspace(-5, 5, 100)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
x = [1,2,3]
y = [5,7,4]
plt.bar(x, y, width = 0.2)
```

```python
x = np.linspace(-5, 5, 20)
y = np.linspace(-5, 5, 20)
X,Y = np.meshgrid(x,y)  #创建网格
Z = (1-X/2 + X**3 + Y**4)*np.exp(-X**2-Y**2)   #Z的函数表达式
plt.contourf(X, Y, Z)
```

```python
x = np.linspace(-5, 5, 20)
y = np.linspace(-5, 5, 20)
X,Y = np.meshgrid(x,y)  #创建网格
Z = (1-X/2 + X**3 + Y**4)*np.exp(-X**2-Y**2)   #Z的函数表达式
plt.pcolormesh(X, Y, Z)  #用于创建具有非规则矩形网格的伪彩色图
```
