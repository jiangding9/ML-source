```python
from pandas import read_csv
```

```python
with open('knn.csv', encoding = 'UTF-8') as f:
    names = ['x', 'y', 'labels']
    data = read_csv(f, names = names)
data
```

```python
#取x这一列，然后取这一列的前200行元素，逗号后面一般用于取列的数目
print(data['x'].iloc[:200,])
```

```python
#取x这一列，然后取这一列的前200行元素，逗号后面一般用于取列的数目
print(data.iloc[1,])
```

```python
import matplotlib.pyplot as plt

x1 = data['x'].iloc[:200,]  #iloc:
y1 = data['y'].iloc[:200,]
x2 = data['x'].iloc[200:400,]
y2 = data['y'].iloc[200:400,]
x3 = data['x'].iloc[400:600,]
y3 = data['y'].iloc[400:600,]

plt.figure(figsize = (6,6))
plt.scatter(x1, y1, c = 'b', marker = 's', s =50)
plt.scatter(x2, y2, c = 'r', marker = '^', s =50)
plt.scatter(x3, y3, c = 'g', s =50)
```

```python
keep = ['x', 'y']
X = data[keep]
y = data['labels'].values
print(X)
print(y)
```

```python
from sklearn import tree

clf = tree.DecisionTreeClassifier()
clf.fit(X, y)
```

```python
import numpy as np
xx, yy= np.meshgrid(np.arange(0.2, 1.31, 0.01), np.arange(0.2, 1.31, 0.01))
```

```python
x0, y0 = np.meshgrid(np.arange(0, 5, 1), np.arange(0, 5, 1))
print(x0)
print(y0)
plt.scatter(x0, y0)
```

```python
coords = np.stack((xx.reshape(-1), yy.reshape(-1)), axis =1)
print(xx)
print(yy)
print(xx.shape)
print(yy.shape)
print(coords)  #注意这里，xx的112*112数组，yy也是。但是在stack之后，得到的是一个[112*112,2]的数组
print(coords.shape)
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(np.stack((a, b), axis = 1)) #axis=1表示按列排列，=0时表示按行排列
```

```python
Z = clf.predict(coords)
print(xx.shape)
print(yy.shape)
print(Z.shape)
Z = Z.reshape(xx.shape)   #Z得到后，把点在图上可视化
```

```python
Z
```

```python
from matplotlib.colors import ListedColormap

light_rgb =  ListedColormap(['#AAAAFF', '#FFAAAA', '#AAFFAA'])  #调整颜色

plt.figure(figsize = (6,6))

plt.pcolormesh(xx, yy, Z, cmap = light_rgb)  #应用颜色
 
plt.scatter(x1, y1, c = 'b', marker = 's', s =50, alpha = 0.8) #alpha= 设置透明度
plt.scatter(x2, y2, c = 'r', marker = '^', s =50, alpha = 0.8)
plt.scatter(x3, y3, c = 'g', s =50, alpha = 0.8) 
```

```python
from sklearn import neighbors

clf2 = neighbors.KNeighborsClassifier(30)  #30的意思：每个测试点为圆心寻找最近的30个点来投票判断其属于什么类别。
clf2.fit(X, y)
```

```python
Z2 = clf2.predict(coords)
Z2 = Z2.reshape(xx.shape)
```

```python
from matplotlib.colors import ListedColormap

light_rgb =  ListedColormap(['#AAAAFF', '#FFAAAA', '#AAFFAA'])  #调整颜色

plt.figure(figsize = (6,6))

plt.pcolormesh(xx, yy, Z2, cmap = light_rgb)  #应用颜色
 
plt.scatter(x1, y1, c = 'b', marker = 's', s =50, alpha = 0.8) #alpha= 设置透明度
plt.scatter(x2, y2, c = 'r', marker = '^', s =50, alpha = 0.8)
plt.scatter(x3, y3, c = 'g', s =50, alpha = 0.8) 
```
