```python
z = np.array(df['z']) * 1e10
lnI = np.log(np.array(df['Imean']) * 1e12) #np.log求对数
lnI
```

```python
#画出lnI与z的关系图
import numpy as np
import matplotlib.pyplot as plt

plt.scatter(z, lnI)  #画图
```

```python
fit_res = np.polyfit(z, lnI, 1)  #x = z, y = lnI，以1次方进行多元拟合（即线性拟合）
print(fit_res)
```

所以上面的两个输出量分别是斜率和截距

实操：谱数据的平滑 

```python
with open('UV_noise.csv', mode = 'r') as f:  #这里不用写encoding = 'UTF-8'
    data = np.loadtxt(f, delimiter = ',')   #delimiter指定分隔符
data
```

```python
data.shape
```

```python
data[0,1]  #输出0列，1行的元素？
```

```python
data = data.T  #取吸收值进行平滑，因此先对数据进行转置
data
```

```python
wl = data[0]  #取波长
Abs = data[1]   #取吸收
```

```python
from scipy.signal import savgol_filter    #savgol_filter是一个数据平滑的函数,savitzky-golay
Abs_smooth = savgol_filter(Abs, 9, 2)       #进行数据平滑, 方法是取9个点进行2次函数拟合。
```

csv文件的常规写法

```python
with open('UV.csv', mode = 'a', encoding = 'UTF-8') as f:
    for i in range(len(wl)):
        f.write(str(wl[i]) + ',' + str(Abs[i]) + ',' + str(Abs_smooth[i]) + '\n')  #把所有东西写进去
```
